# Projekt VVP - bludiště


## Co projekt obsahuje ?
### a) Class Maze
Classa Maze reprezentuje vytvořené nebo načtené maticové bludiště, když s ní chceme pracovat musíme si uvědomit:
1) Když vytvoříme classu Maze je nutné zadat správné parametry. Jsou dvě základní možnosti pro vytvoření Maze a) "create" b) "load". Tyto dvě možnosti se davájí na pozici prvního argumentu. Možnost "create" vytváří vlastní budiště ze šablony a možnost "load" načítá předem vytvořené bludiště ze souboru.
2) Pokud zvolíme možnost "create" musíme doplnit ještě argumenty size(velikost našeho chtěného bludiště) a template(šablona bludiště), size je u "create" povinný argument a template není, protože pokud jej nespecifikujeme, tak defaultně je nastavená šablona "empty".
3) Pokud zvolíme "load" tak potřebujeme specifikovat soubor, ze kterého bludiště budeme brát, to pomocí argumetnu file, jiné argumenty nás nezajímají. 

### b) Další funkce, které budeme využívat
1) Shortest_path: Najde nejkratší cestu v bludišti z levého horního rohu do pravého dolního rohu a vrací seznam bodů, kterým tato cesta prochází.
2) Maze_picture: Vykreslí dané bludiště s cestou, která je zadána.


#### V této buňce tvoříme bludiště se šablonou "smiley" o velikosti 50x50, poté najdeme nejkratší cestu a bludiště s nějkratší cestou vykreslíme.

```python
from projekt import Maze

maze = Maze("create",50,"smiley")
path = Maze.shortest_path(maze)
maze.maze_picture(path)
```

#### V této buňce načítáme bludiště ze souboru, poté zase hledáme nejkratší cestu a zase vykreslíme

```python
maze_instance = Maze("load",file="data//maze_5.csv")
path = maze_instance.shortest_path()
maze_instance.maze_picture(path)
```
